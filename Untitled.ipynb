```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

 # Concoide de Nicomedes



$x = a + cos(t)$ <br\>
$y = a*tan(t)+sin(t)$

```python

fig = plt.figure()
ax = plt.axes(xlim=(-4,4), ylim=(-4,4))
line, = ax.plot([], [], lw=2)
t = np.arange(0, 2*np.pi, 0.01) 

line2, = ax.plot([], [], lw=2,linestyle=':',color="black")

def init():
    line.set_data([], [])
    line2.set_data([],[])
    return line,line2,

def animate(i):
    x = (-1.5+i/200.0)+np.cos(t)
    y = (-1.5+i/200.0)*np.tan(t)+np.sin(t)
    x1 = -1.5+i/200.0
    x2 = np.arange(-4,4,0.01)
    line.set_data(x, y)
    line2.set_data(x1,x2)
    return line,line2,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=600, interval=30, blit=True)
HTML(anim.to_html5_video())
```

# Cicloide


$x = r(\theta sin(\theta))$ <br\>
$y = r(1-cos(\theta))$

```python
fig2 = plt.figure()
ax = plt.axes(xlim=(-0,20), ylim=(0,12))
line, = ax.plot([], [], lw=2)
line2, = ax.plot([], [], lw=2)
r=1
t = np.linspace(0, 12*np.pi, 400)
"x1 = r*(t-np.sin(t))"
"x2 = r*(1-np.cos(t))"
"plt.plot(x1,x2)"
a = np.arange(0, 2*np.pi, 0.01) 
b = np.arange(-8*np.pi,0, 0.01)

def init():
    line.set_data([], [])
    line2.set_data([],[])
    return line,line2

def animate(i):
    x = r*(i/10.0)+np.cos(a)
    y = r+np.sin(a)
    X = r*(b+i/10.0-np.sin(b+i/10.0))
    Y = r*(1-np.cos(b+i/10.0))
    line.set_data(x,y)
    line2.set_data(X,Y)
    return line,line2

ani = animation.FuncAnimation(fig2, animate, init_func=init,
                               frames=200, interval=50, blit=True)
HTML(ani.to_html5_video())
```
